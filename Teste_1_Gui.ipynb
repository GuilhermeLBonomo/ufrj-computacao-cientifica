### Questão 4
Escreva um trecho que código Python que imprima de maneira formatada (bonitinha) as oito primeiro linhas do triângulo de Pascal.

```python
from math import factorial


def triangulo_pascal(n: int) -> None:
    for i in range(n):
        for x in range(n-i+1):
            print(' ',end=' ')
        for x in range(i+1):
            numero:int = factorial(i)//(factorial(x)*factorial(i-x))
            print(f'{numero:^4}', end=" ")
        print(' ')


triangulo_pascal(8)
```

```python
from math import log10


LEI_BENFORD:dict[int,float] = {d:round(log10((d+1)/d),3) for d in range(1,10)}

#Otimizei essa função pq achei que era ela o problema mas foi um erro no código que tava fazendo demorar...
def fib(n: int):
    a, b = 0, 1
    for _ in range(n if n > 1 else 0):
        a, b = b, a + b
    return a


def fiblist(num:int) -> list[int]:
    return [fib(x) for x in range(num)]


def get_d(num:int) -> int:
    return int(str(abs(num))[0])


def count_d(seq: list[int]) -> dict[int,int]:
    dist:dict[int,int] = {d:0 for d in range(1, 10)}
    for num in seq:
        dist[get_d(num)] += 1
    return dist


def get_benford_dist(seq: list[int]) -> tuple[dict[int, int], dict[int, float]]:
    dist: dict[int, int] = {d: 0 for d in range(1, 10)}
    for num in seq:
        if get_d(num) == 0:
            continue
        dist[get_d(num)] += 1
    TOT: int = sum(dist.values())
    benford_dist: dict[int, float] = {
        d: (count / TOT) * LEI_BENFORD[d] for d, count in dist.items()}
    return dist, benford_dist


# print(fiblist(6))
# print(get_d(172626))
# print(benford(1223))

def main(x=500, y=1000):
    f500: list[int] = fiblist(x)
    f1k: list[int] = fiblist(y)
    b500: tuple[dict[int, int], dict[int, float]] = get_benford_dist(f500)
    b1k: tuple[dict[int, int], dict[int, float]] = get_benford_dist(f1k)

    d500, bd500 = b500
    d1k, bd1k = b1k

    print("Distribuição dos primeiros dígitos para N=500:")
    print("Dígito | Fibonacci | Benford")
    for d in range(1, 10):
        print(f"   {d}   |   {d500[d]}    |   {bd500[d]:.4f}")

    print("\nDistribuição dos primeiros dígitos para N=1000:")
    print("Dígito | Fibonacci | Benford")
    for d in range(1, 10):
        print(f"   {d}   |   {d1k[d]}    |   {bd1k[d]:.4f}")

main()
```

### Questão 6
Escreva um loop usando `while` para calcular a [média aritmética-geometrica](https://en.wikipedia.org/wiki/Arithmetic%E2%80%93geometric_mean) (AGM) de dois números reais positivos, $x$ e $y$, definida como o limite das sequência:

$$
\begin{align}
a_{n+1} &= \frac{1}{2} (a_n + g_n), \\
g_{n+1} &= \sqrt{a_n g_n},
\end{align}
$$

começando com $a_0 = x$ e $g_0 = y$. Ambas as sequências convergem para o mesmo número, denotado por $\textrm{agm}(x,y)$. Use sua implementação para calcular a constante de *Gauss*, $G = 1/\textrm{agm}(1,\sqrt{2})$.

**Importante:** Você terá que usar um critério de parada para iteração.

```python
def agm(x:float, y:float, precisao=1e-9, lim_inter=1e9) -> float:
    a, g = x, y
    while (abs(a-g) > precisao) and lim_inter > 0:
        a_n1: float = (a+g)/2
        g_n1: float = (a*g)**0.5
        a = a_n1
        g = g_n1
        lim_inter -= 1
    return a


G: float = 1/(agm(1, (2**0.5)))

print(f'Constante de Gauss: {G}')
```
